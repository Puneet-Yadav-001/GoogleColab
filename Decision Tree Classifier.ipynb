```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_excel('shows.xlsx')
df.head()
```

```python
# mapping non numerical val
print(df['Nationality'].unique())
print(df['Go'].unique())
```

```python
contry = {'UK':0, 'USA':1, 'N':2}
tar = {'NO':0,'YES':1}
df['Nationality'] = [contry[i] for i in df.Nationality]
df['Go'] = [tar[i] for i in df.Go]
df.head()
```

```python
X = df.drop(['Go'],axis =1)
y = df.Go
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
print(dtree.predict([[40, 10, 7, 1]]))
```
